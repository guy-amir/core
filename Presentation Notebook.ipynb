```python
cnn = pd.read_csv("./treeFalselr0.01.csv") #CNN w/o tree
cnn
```

```python
acc = pd.DataFrame()
acc['val_cnn'] = cnn.val_acc_list
acc['val_d5'] = d5.val_acc_list
acc['val_d12'] = d12.val_acc_list
acc['train_cnn'] = cnn.train_acc_list
acc['train_d5'] = d5.train_acc_list
acc['train_d12'] = d12.train_acc_list
```

```python
acc
```

```python
acc.plot(figsize=(15,10))
```
